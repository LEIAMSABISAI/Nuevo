### REGRESOR LINEAL

La regresion lineal es un tipo de algoritmo de Machine Learning basado en el aprendizaje supervisado. Este algoritmo precide el valor de una variable dependiente(Y) y una variable independiente (X), esto lo transforma a una funcion que pueda atravesar estos puntos con el minimo error posible.

El modelo de regresión lineal simple lo podemos representar con la expresión:

$$\hat{y}= \alpha + \beta X + \epsilon $$
    
Donde:

$\alpha$ es el origen que sucede en Y cuando X vale 0

$\beta$ es la pendiente de la recta y nos indica el cambio de Y con respecto a X

$\epsilon$ es el error que tenemos de acuerdo a conjunto de datos dados.


Para efectos de la practica usaremos teta en nuestra ecuacion, quedando de la siguiente manera:

$$\hat{y}=\theta_0 + \theta_1 x_1$$



### BATCH GRADIENT DESCENT

Continuando en buqueda de un algoritmo capaz de obtener un modelo de regresion lineal no encontramos con el Gradient Descent, el cual lo podemos dividir en el Batch Gradient Descent y el Stochastic Gradient Descent (del cuál hablaremos mas tarde), en el primero usamos todo el conjunto de datos para aproximar el modelo lo mas exacto posible, este tipo de algoritmo nos permite tener una gran exactitud, sin embargo, al tomar todo el conjunto de datos, esto nos representa un incremento considerable en los requisitos de computo y el tiempo necesario para lograrlo.

Este modelo usaremos la expresión:

$\bigtriangledown_{\theta}$ MSE = 2/m X$^{T}$ (X$\theta$ - Y)

Y a su vez tendremos que:

$\theta_{new}$ = $\theta$-n$\bigtriangledown_{\theta}$MSE





### STOCHASTIC GRADIENT DESCENT

Cuando el tiempo, equipo de computo, es limitado, o bien el numero de datos es demasiado grande para optar por el algoritmo de Batch, entonces procedemos a utilizar el Stochastic, este algoritmo usa la misma expresion de Gradient Descent con la diferencia que los datos ingresados para su aprendizaje son completamente aleatorios, por lo cual la precision que podremos obtener será mucho menor que con Bacth, sin embargo el tiempo se reduce considerablemente, sin emabrgo, en ocasiones sera necesario entrenar varias veces el modelo, debido a que los datos aleatorios podrian generar un sistema algo inestable o "estocástico".

En este modelo volveremos a usar la expresión:

$\bigtriangledown_{\theta}$ MSE = 2/m X$^{T}$ (X$\theta$ - Y)

Y a su vez:

$\theta_{new}$ = $\theta$-n$\bigtriangledown_{\theta}$MSE




### MINI BATCH GRADIENT DESCENT

Como punto intermedio podemos obtar por tomar un cierto porcentaje de los datos, en este caso entre un 10% y 20%, esto nos permite tener una aproximacion mas cercana sin tener que aplicar todos el tiempo y los requisitos de computo que nos implica el Batch completo, al tratarse del mismo Gradient Descent usaremos las mismas expresiones con la diferencia de los datos usados.

$\bigtriangledown_{\theta}$ MSE = 2/m X$^{T}$ (X$\theta$ - Y)

Y a su vez:

$\theta_{new}$ = $\theta$-n$\bigtriangledown_{\theta}$MSE

### BATCH GRADIENT DESCENT

```python
import numpy as np #Importamos la libreria de numpy para trabajar con matrices
import matplotlib.pyplot as plt #Importamos la libreria de matplotlib para trabajar gráficas

# A continuacion creamos datos random para X,Xb y Y
X=2*np.random.rand(100,1)
Xb=np.c_[np.ones((100,1)),X]
y=4+3*X+np.random.rand(100,1)

#gráficamos los puntos
plt.plot(X,y,'*g')
```

Creamos las expresiones necesarias para realizar el modelo

```python
#init
eta=0.01
nIter=500
m=100
theta=np.random.rand(2,1)
for iterations in range(nIter):
    grad=2/m*Xb.T.dot(Xb.dot(theta)-y)
    theta=theta-eta*grad
print(theta)
```

creamos las matrices necesarias

```python
Xnew=np.array([[0],[2]])
Xnewb=np.c_[np.ones((2,1)),Xnew]
Xnewb
yPred=Xnewb.dot(theta)
yPred
```

Y aplicamos la funcion para graficar punto por punto

```python
def PlotGradientDescent(theta,eta):
    m=len(Xb)
    plt.plot(X,y,'.b')
    nIter=500
    for iteration in range (nIter):
        if iteration<50:
            yPred=Xnewb.dot(theta)
            style='-r' if iteration>0 else 'r--'
            plt.plot(Xnew,yPred,style)
        grad=2/m*Xb.T.dot(Xb.dot(theta)-y)
        theta=theta-eta*grad
    plt.xlabel("$x_1$",fontsize=18)
    plt.axis([0,2,0,15])
    plt.title(r"$\eta={}$".format(eta),fontsize=16)
```

Por ultimo mandamos llamar las funciones

```python
#init
np.random.seed(42)
theta=np.random.rand(2,1)
#plots
plt.figure(figsize=(10,4))
plt.subplot(131); PlotGradientDescent(theta, eta=0.02)
plt.subplot(132); PlotGradientDescent(theta, eta=0.1)
plt.subplot(133); PlotGradientDescent(theta, eta=0.45)
```

### Stochastic Gradient Descent

Primero creamos un conjunto de datos a partir de una semilla prediseñada

```python
m=len(Xb)
np.random.seed(50)
nEpoch=50
t0, t1=5, 50
theta=np.random.randn(2,1)
```

```python
def LearninSchedule(t):
    return t0/(t+t1)
```

Creamos una función donde grafiquemos punto por punto y obtenemos el resultado

```python
for epoch in range(nEpoch):
    for i in range(m):
        if epoch==0 and i<25:
            yPred=Xnewb.dot(theta)
            style=':y' if i>0 else '--r'
            plt.plot(Xnew,yPred,style)
        ranIndex=np.random.randint(m)
        xi=Xb[ranIndex:ranIndex+1]
        yi=y[ranIndex:ranIndex+1]
        grad=2*xi.T.dot(xi.dot(theta)-yi)
        eta=LearninSchedule(epoch*m+i)
        theta=theta-eta*grad
            
plt.plot(X,y,'.b')
```
